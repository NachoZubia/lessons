```python
def change_value(x):
    if 'M' in str(x):
        res= float(x.replace("M",""))*1000000.0
    elif 'K' in str(x):
        res = float(x.replace("K",""))*1000.0
    else:
        res = float(x)
    return res

```

## Separamos las columnas con sumas en los valores y posteriormente concatenamos dataframes y eliminamos el original

```python
col_suma = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF',
       'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK']
new_df = pd.DataFrame(columns=col_suma)
for i in col_suma:
    last_df = df[i].str.split('+', expand = True).rename(columns={0:f'{i}(-)', 1:f'{i}(+)'})
    new_df = pd.concat([new_df, last_df], axis=1)
new_df.drop(col_suma, axis=1, inplace=True)
```

```python
new_df.columns
```

```python
df = df.drop(df[df['Hits'].str.contains('K')].index, axis=0)
```

```python
df.drop(col_suma, axis=1, inplace=True)
```

```python
df = pd.concat([df, new_df], axis=1)
```

```python
df.head(2)
```

## Importamos librerías y dividimos el dataframe en Test y Train

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
y = df['OVA']
```

```python
X = df.drop(['OVA'], axis= 1)
```

```python
y.dropna(inplace=True)
X.dropna(inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) #dividimos el modelo en train y test
```

## Modelo de Regresión Lineal

```python
lm = LinearRegression() #modelo
```

```python
lm.fit(X_train, y_train) #model train
```

```python
y_predict = lm.predict(X_test) #model prediction
```

## Regularización

```python
from sklearn.linear_model import Lasso       # L1
from sklearn.linear_model import Ridge       # L2
from sklearn.linear_model import ElasticNet  # L1+L2
```

```python
# Lasso L1

lasso=Lasso()
lasso.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
```

```python
# Ridge L2

ridge=Ridge()
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)
```

```python
# ElasticNet L1+L2

elastic=ElasticNet()
elastic.fit(X_train, y_train)

y_pred_elastic = elastic.predict(X_test)
```

## R2

```python
train_score=lm.score(X_train, y_train) #R2
test_score=lm.score(X_test, y_test)
model_name = ['Regresión Lineal (lm)']
print ('Model: {}, train R2: {} -- test R2: {}'.format(model_name, train_score, test_score))
```

## MSE

```python
from sklearn.metrics import mean_squared_error as mse
train_mse=mse(lm.predict(X_train), y_train) #MSE
test_mse=mse(y_predict, y_test)
print ('Model: {}, train MSE: {} -- test MSE: {}'.format(model_name, train_mse, test_mse))
```

## RMSE

```python
train_rmse=mse(lm.predict(X_train), y_train)**0.5 #RMSE
test_rmse=mse(y_predict, y_test)**0.5
print ('Model: {}, train RMSE: {} -- test RMSE: {}'.format(model_name, train_rmse, test_rmse))
```

## MAE

```python
from sklearn.metrics import mean_absolute_error as mae
train_mae=mae(lm.predict(X_train), y_train) #MAE
test_mae=mae(y_predict, y_test)
print ('Model: {}, train MAE: {} -- test MAE: {}'.format(model_name, train_mae, test_mae))
```

<font color='blue'>
Obtenemos un MAE de valor 1.55 lo que significa que obtendremos un "OVA" para nuestros jugadores con un error absoluto de +- 1.55 su valoración real
</font>

## Utilizando modelo Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor as RFR

rfr=RFR()
rfr.fit(X_train, y_train)

y_pred_rfr = rfr.predict(X_test)
```

## MAE

```python
train_mae=mae(rfr.predict(X_train), y_train) #MAE
test_mae=mae(y_pred_rfr, y_test)
model_name_rfr = ['Random Forest Regressor']
print ('Model: {}, train MAE(RFR): {} -- test MAE(RFR): {}'.format(model_name_rfr, train_mae, test_mae))
```

<font color='blue'>
Con el fin de optimizar la solución probamos a testar el modelo con un modelo de regresión no lineal más complejo basado en árboles de decisión, Random Fores Regressor, con ello hemos conseguido optimizar el error del modelo en más de 1.1 puntos absolutos a partir del MAE, con un error final de MAE(RFR) = 0.323
</font>
